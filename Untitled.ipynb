```python
# configs
import hydra
hydra.core.global_hydra.GlobalHydra.instance().clear()

agent_name = "mae_sep_seg2_add"
mode = "test"
train_n_demos = 1000
checkpoint_type = "test_best"
exps_name = "exps_extra_sep_seg2_add"

tasks=["assembling-kits-seq-full",
    "packing-boxes-pairs-full",
    "stack-block-pyramid-seq-full",
    "separating-piles-full",
    "towers-of-hanoi-seq-full",
    "put-block-in-bowl-full",
    "packing-seen-google-objects-group",
    "packing-unseen-google-objects-group",
    "packing-seen-google-objects-seq",
    "packing-unseen-google-objects-seq"]

eval_task = tasks[9]
initialize(config_path='cliport/cfg', job_name="test_app")
vcfg=compose(config_name='eval', overrides=[f"model_task=multi-language-conditioned", 
                                            f"eval_task={eval_task}", 
                                            f"agent={agent_name}",
                                            f"mode={mode}",
                                            f"n_demos={100}",
                                            f"train_demos={train_n_demos}",
                                            f"exp_folder={exps_name}",
                                            f"checkpoint_type={checkpoint_type}",
                                            f"update_results=False"])
tcfg = utils.load_hydra_config(vcfg['train_config'])
name = '{}-{}-n{}'.format(eval_task, vcfg['agent'], vcfg['n_demos'])
```

```python
# dataset 
from torch.utils.data import DataLoader, Subset
ds = dataset.RavensDataset(os.path.join(vcfg['data_dir'], f"{eval_task}-{mode}"),
                                   tcfg,
                                   n_demos=vcfg['n_demos'],
                                   augment=False)
val_ds = DataLoader(ds, batch_size=1, shuffle=False, pin_memory=True, num_workers=0)
val_ds = iter(val_ds)
# ckpt
ckpt_manager = CkptManager(vcfg)
eval_pick, eval_place = ckpt_manager.get_test_ckpt()
```

```python
# load trained agent
tcfg['pretrain_path'] = None 
tcfg['train']['batchnorm'] = True
agent = agents.names[agent_name](name, tcfg, None, ds, 'both')
agent.eval()

existing_results = {}
model_pick, model_place, test_name, to_eval = get_model_path(
            eval_pick[0], eval_place[0], vcfg, existing_results)

# model load checkpoint
agent.load_sep(model_pick, model_place)
print(f"Loaded: {eval_pick[0]}, {eval_place[0]}")
```

```python
import torch
from torchvision import transforms
from PIL import Image
def load_image_to_tensor(image_path):
    """
    Load an image from the specified path and convert it to a PyTorch tensor.

    Args:
        image_path (str): The path to the image file.

    Returns:
        torch.Tensor: The image converted to a PyTorch tensor.
    """
    # Define the transform to convert the image to a tensor
    transform = transforms.Compose([
        transforms.Resize((320, 160)),
        transforms.ToTensor()
    ])

    # Load the image
    image = Image.open(image_path)

    # Apply the transform to the image
    image_tensor = transform(image)

    # Optional: Move the tensor to GPU if available
    if torch.cuda.is_available():
        image_tensor = image_tensor.to('cuda')
        print('Tensor moved to GPU')
    
    return image_tensor

```
